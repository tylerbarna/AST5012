```python
import numpy as np
import astropy
import scipy
import matplotlib.pyplot as plt

from astropy import constants as const
from astropy import units as u
```

## Part a

```python
## Carbon Cooling Function (Eqn. 1.37) 
CCool = lambda T: 5.2E-27  * (T/100)**(0.13) * np.exp(-91.2/T) * u.erg * (u.cm)**3 * (u.s)**(-1)
```

```python
## Oxygen Cooling Function (Eqn. 1.38)
OCool = lambda T: 4.1E-27  * (T/100)**(0.42) * np.exp(-228/T) * u.erg * (u.cm)**3 * (u.s)**(-1)
```

```python
## Lyman Alpha Cooling Function (Eqn. 1.38)
AlphaCool = lambda T: 6.0E-22  * (T/1e4)**(-0.5) * np.exp(-1.18e5/T) * u.erg * (u.cm)**3 * (u.s)**(-1)
```

```python
Temps = np.linspace(1e1,2e4,1e5)
fig,ax = plt.subplots(figsize=(10,5))
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid()
ax.set_xlabel("Temperature [K]")
ax.set_ylabel(r"Λ [erg cm$^3$ s$^{-1}$]")
ax.set_ylim(1e-30,1e-24)

ax.plot(Temps, CCool(Temps),label="[CII]")
ax.plot(Temps, OCool(Temps),label="[OI]")
ax.plot(Temps, AlphaCool(Temps),label="[Lyα]")
ax.plot(Temps, CCool(Temps)+OCool(Temps)+AlphaCool(Temps),label="Total")

ax.legend();
```
